```python
'''8-5 ⭐️ Посмотрим на данные за период с апреля по декабрь.
Объедините сэмплы данных за нужные месяцы в один общий датасет bikes. Сделайте преобразование по дням для каждой группы 
пользователей (usertype), затем выберите дни, в которые число аренд, сделанных customers, было больше, чем у subscribers.
Данные за 
Q2: (1) апрель, (2) май, (3) июнь
Q3: (4) июль, (5) август, (6) сентябрь
Q4: (7) октябрь, (8) ноябрь, (9) декабрь'''
```

```python
bikes = pd.concat([
           pd.read_csv("bikes_q2_sample_apr.csv"), 
           pd.read_csv("bikes_q2_sample_jun.csv"), 
           pd.read_csv("bikes_q2_sample_may.csv"), 
           pd.read_csv("bikes_q3_sample_aug.csv"),
           pd.read_csv("bikes_q3_sample_july.csv"),
           pd.read_csv("bikes_q3_sample_sep.csv"),
           pd.read_csv("bikes_q4_sample_dec.csv"),
           pd.read_csv("bikes_q4_sample_nov.csv"),
           pd.read_csv("bikes_q4_sample_oct.csv"),
            ])
```

```python
bikes[['start_time', 'end_time']] = bikes[['start_time', 'end_time']].apply(pd.to_datetime)
```

```python
bikes = bikes.sort_values('start_time').set_index('start_time')
```

```python
bikes_slice = bikes.groupby('usertype').resample(rule = 'D').size().T
bikes_slice
```

```python
bikes_slice.query('Customer > Subscriber')
```

```python
'''8-6 Отлично! Следующий шаг – более подробно посмотреть на летний период.
Еще один плюс использования дат в качестве индексов – возможность выбрать данные за интересующий нас промежуток времени. 
В переменную bikes_summer сохраните наблюдения с 1 июня по 31 августа. Затем запишите в top_destination наиболее популярный 
пункт назначения (его название). Агрегируйте данные по дням и определите, в какой день в полученный пункт (top_destination) 
было совершено меньше всего поездок. Дату сохраните в bad_day, отформатировав timestamp с помощью .strftime('%Y-%m-%d').'''
```

```python
bikes_summer = bikes.loc['2018-06-01':'2018-08-31']
bikes_summer.head()
```

```python
top_destination = bikes_summer.to_station_name.value_counts().idxmax()
top_destination
```

```python
bad_day = bikes_summer.query('to_station_name == @top_destination').resample(rule = 'D').size().idxmin().strftime('%Y-%m-%d')
bad_day
```

```python
'''8-7 Куда больше всего ездят на выходных? Туда же, куда и в будние дни, или в другие пункты назначения?
Используя данные за период с 1 июня по 31 августа, выберите верные утверждения.

'''
```

```python
bikes_summer['weekday'] = bikes_summer.index.weekday<5
```

```python
bikes_summer['day_name'] = bikes_summer.index.day_name()
bikes_summer.head()
```

```python
bikes_summer.groupby(['weekday', 'day_name']) \
        .to_station_name \
        .value_counts() \
        .nlargest(10)
```

```python
bikes_summer.query('weekday==True')\
        .to_station_name \
        .value_counts() \
        .nlargest(10)
```
