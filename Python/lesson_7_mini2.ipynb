```python
'''
В этом задании нужно написать цикл, в котором сформировать датафрейм с информацией об объектах. То есть: 
создадим датафрейм и в каждой итерации цикла будем добавлять в него строку с информацией об объекте. 
Просто посчитать, сколько объектов относятся к японской культуре

Чтобы выполнить задание необходимо: 

1. Создать пустой датафрейм 
2. Написать цикл, который будет перебирать объекты из сохраненного ранее `sunflower_ids`
3. Написать запрос к API, вместо `object_id` подставить **id** из `sunflower_ids` и сформировать строчку в датафрейме для объекта
4. В созданном датафрейме посчитать сколько объектов относятся к японской культуре
5. Может понадобиться конструкция `try...except`

Убедитесь, что вы правильно сформировали датафрейм и указали количество объектов относящихся к Японской культуре 
`culture == 'Japan’`'''
```

```python
cols=['objectID', 'title', 'artistDisplayName', 'department', 'objectBeginDate', 'objectEndDate',
'period', 'objectName', 'culture']

df_flower = pd.DataFrame(columns=cols)
```

```python
for object_id in sunflower_ids:
    try:
        r_test = requests.get(f'https://collectionapi.metmuseum.org/public/collection/v1/objects/{object_id}')
        obj_test = r_test.json()
        df_temporary = pd.DataFrame([[obj_test['objectID'],
                                      obj_test['title'],
                                      obj_test['artistDisplayName'],
                                      obj_test['department'],
                                      obj_test['objectBeginDate'],
                                      obj_test['objectEndDate'],
                                      obj_test['period'],
                                      obj_test['objectName'],
                                      obj_test['culture']]],
                                      columns=cols)
        df_flower = pd.concat([df_flower, df_temporary])
    except:
        print(f'Error{object_id}')
```

```python
'''
В задании нужно написать запрос к API с несколькими параметрами: с departmentId = 6 ("Asian art") и тегом "cat". 
Сохранить objectIDs в переменную cat_ids. Пример обращения к поиску :

r = requests.get('https://collectionapi.metmuseum.org/public/collection/v1/search?departmentId=1&q=cat')
cats = r.json()

Далее, написать цикл как в предыдущем задании, в котором сформировать датафрейм с информацией об первой 1000 
объектов из списка. В качестве ответа укажите, в каком периоде (period) было выполнено наибольшее количество работ

    
    Напишите запрос к API с параметрами `departmentId = **6**` ("Asian art") и `q="cat"` и сохраните `objectIDs` 
    в переменную `cat_ids`. Далее, создайте пустой датафрейм и напишите цикл, который будет перебирать первую 
    1000 объектов из сохраненного ранее `cat_id`. Затем напишите запрос к API, вместо `object_id` подставьте **id** 
    из `cat_id` и сформируйте строчку в датафрейме для объекта. Осталось посчитать в каком периоде (`period`) было 
    выполнено наибольшее количество работ, используя `value_counts()`
'''
```

```python
r = requests.get('https://collectionapi.metmuseum.org/public/collection/v1/search?departmentId=6&q=cat')
cats = r.json()
cat_ids = []
for i in cats:
    cat_ids.append(cats['objectIDs'])
cat_ids=cat_ids[0]
cat_ids
```

```python
cols=['objectID', 'title', 'artistDisplayName', 'department', 'objectBeginDate', 'objectEndDate',
'period', 'objectName', 'culture']

df_cats = pd.DataFrame(columns=cols)
```

```python
for object_id in cat_ids[:1000]:
    try:
        r_test = requests.get(f'https://collectionapi.metmuseum.org/public/collection/v1/objects/{object_id}')
        obj_test = r_test.json()
        df_temporary_cats = pd.DataFrame([[obj_test['objectID'],
                                      obj_test['title'],
                                      obj_test['artistDisplayName'],
                                      obj_test['department'],
                                      obj_test['objectBeginDate'],
                                      obj_test['objectEndDate'],
                                      obj_test['period'],
                                      obj_test['objectName'],
                                      obj_test['culture']]],
                                      columns=cols)
        df_cats = pd.concat([df_cats, df_temporary_cats])
    except:
        print(f'Error {object_id}')
```
